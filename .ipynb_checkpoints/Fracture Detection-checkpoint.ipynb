```python
# Imports
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

import os, warnings, shutil, json, zipfile, random, math, cv2
import matplotlib.pyplot as plt
from matplotlib import gridspec
from PIL import Image

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
from keras.applications.vgg16 import VGG16
from keras.applications.resnet50 import ResNet50
from keras.applications.resnet import ResNet152
from sklearn.metrics import confusion_matrix
```

```python
#loading data

def convert_to_float(image, label):
    image = tf.image.convert_image_dtype(image, dtype=tf.float32)
    return image, label

size = 100

train_d = image_dataset_from_directory('train', image_size = [size, size], interpolation = 'bilinear')
test_d = image_dataset_from_directory('test', image_size = [size, size], interpolation = 'bilinear')
valid_d = image_dataset_from_directory('val', image_size = [size, size], interpolation = 'bilinear')

AUTOTUNE = tf.data.experimental.AUTOTUNE
train = (train_d.map(convert_to_float).cache().prefetch(buffer_size=AUTOTUNE))
test = (test_d.map(convert_to_float).cache().prefetch(buffer_size=AUTOTUNE))
valid = (valid_d.map(convert_to_float).cache().prefetch(buffer_size=AUTOTUNE))
```

```python
#some images raised an error so this finds all the corrupted images
def check_image_readability(folder_path):
    file_list = os.listdir(folder_path)
    
    errored_images = []
    correct_images = []

    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)
        try:
            with Image.open(file_path) as img:
                img.load()
            correct_images.append(file_path)
        except Exception as e:
            print(e)
            print(file_path)
            errored_images.append(file_path)
            continue
    
    if errored_images:
        print("There are errored images")
    else:
        print("All images are readable.")
    return errored_images, correct_images

#move corrupted images for safekeeping
def move_images(images_list, to_dir):
    if not os.path.exists(to_dir):
        os.mkdir(to_dir)
    
    for image_path in images_list:
        image_name = os.path.basename(image_path)
        destination_path = os.path.join(to_dir, image_name)
        shutil.copy(image_path, destination_path)
    
    print("Images moved successfully to", to_dir)
```

```python
#checking data usability
x, y = check_image_readability('val/fractured/')
print(x)
```

```python
move_images(x, 'broken/val/not fractured/')
```

```python
model_one = keras.Sequential([
    
    layers.InputLayer(input_shape=[size, size, 3]),

    layers.BatchNormalization(renorm=True), 
    layers.Conv2D(filters=64, kernel_size=3, activation='relu', padding='same'),
    layers.MaxPool2D(),
    
    layers.BatchNormalization(renorm=True),
    layers.Conv2D(filters=128, kernel_size=3, activation='relu', padding='same'),
    layers.MaxPool2D(),
    
    layers.BatchNormalization(renorm=True),                      
    layers.Conv2D(filters=256, kernel_size=3, activation='relu', padding='same'),
    layers.MaxPool2D(),

    layers.BatchNormalization(renorm=True),
    layers.Conv2D(filters=512, kernel_size=3, activation='relu', padding='same'),
    layers.MaxPool2D(),
                              
    layers.BatchNormalization(renorm=True),
    layers.Conv2D(filters=1024, kernel_size=3, activation='relu', padding='same'),
    layers.MaxPool2D(),
    
    layers.BatchNormalization(renorm=True),
    layers.Flatten(),
    layers.Dense(8, activation='relu'),
    layers.Dense(2, activation='sigmoid')
])
```

```python
vgg=VGG16(include_top=False,weights='imagenet',input_shape=(size,size,3))
vgg.trainable=False

model_vgg = keras.Sequential(
    [vgg, layers.Flatten(), layers.Dense(8, activation='relu'), layers.Dense(2, activation='softmax')])
```

```python
def fitting(model, loss = 'sparse_categorical_crossentropy', metrics =['accuracy'], optimiser = 'adam', ep = 25):
    
    model.compile(optimiser, loss = loss, metrics = metrics)
    history = model.fit(train, validation_data = valid, epochs = ep)
    history_frame = pd.DataFrame(history.history)
    history_frame.loc[:, ['loss', 'val_loss']].plot()
    history_frame.loc[:, [metrics[0], 'val_' + metrics[0]]].plot();
```

```python
fitting(model_one)
```

```python
fitting(model_vgg)
```

```python
def cf_matrix(model):
    actual = np.concatenate([y for x, y in valid], axis=0)
    pred = np.argmax(model.predict(valid), axis=-1)
    cm = confusion_matrix(actual, pred)

    # Plot the confusion matrix
    plt.figure(figsize=(4, 4))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
```

```python
cf_matrix(model_one)
```
